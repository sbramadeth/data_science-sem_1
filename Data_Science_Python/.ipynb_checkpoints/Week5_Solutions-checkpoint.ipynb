### Question 1

Write a program that asks the user to enter the total sales for a store for each day of the week. The store is open 7 days a week. The sales amounts should be stored in a list.
Once the data is collected, use a loop to calculate the total sales for the week and display the result.

```python
allSales = []
week = ["Sunday", "Monday", "Tuesday", "Wednesday",
        "Thursday", "Friday", "Saturday"]
for day in range(7):
    dailySales = int(input(f"Please enter sales for {week[day]}"))
    allSales.append(dailySales)

totalSales = 0.0
for sales in allSales:
    totalSales += sales
    
print (totalSales) 
```

### Question 3
Create a program that creates a list and will populate it with the first 40 Fibonacci numbers (this can be done with just three lines of code).
The program should then ask the user to enter an integer value between 1 and 40 to indicate which number in the Fibonacci series they would like to see and the application should display that number. For example, if the user enters 13, the 13th number is 144.

```python
fibSeq = [0, 1]

for num in range(2, 40):
    fibSeq.append(fibSeq[num-1]+fibSeq[num-2])

index = int(input('What index do you wish to see'))

print(fibSeq[index-1])
```

### Question 5

Use a dictionary to store people’s favorite numbers. Think of five names, and use them as keys in your dictionary. Think of a favorite number for each person, and store each as a value in your dictionary. Print each person’s name and their favorite number. For even more fun, poll a few friends and get some actual data for your program.

```python
def person_dict2():
    print("Lets populate a dictionary with data from five people")
    
    Bruce={"number":"10"}
    Dick={"number":"12"}
    Tim={"number":"09"}
    Damian={"number":"2"}
    Jason={"number":"5"}
    
    full_dict = {"Bruce":"2", "Dick":"5", "Damian":"14",
                 "Tim": "21", "Jason":"7"}
    
    for key in full_dict:
        print(f"{key} favourite number is {full_dict[key]}")
    

person_dict2()
```
