### (i) - Write a loop that counts from 0 to 20 (inclusive) using a while loop. Do the same loop but use a for loop and the range() built-in function.


```python
print("while loop:")

count = 0

while count <=20:
    print(count)
    count+=1

print("\n")
print("for loop:")


for count_2 in range(0,20):
    print(count_2)
```

### (ii) - Write a program that will ask the user to enter an integer value greater than 0. If a user enters a negative value the program should continue to ask the user to enter a positive value (use a while loop to achieve this). Once the user has entered a positive value the program should print out all numbers between 0 and the positive value (use a for loop for this)

```python
while True:
    number = input("Please enter an integer greater than zero: ")
    if int(number) >=0:
        for x in range(0, int(number)):
            print(x)
        break
    else:
        print("Error, this program will only accept positive numbers!")
        print("\n")
```

### (iii) - Write a program that will print out the ten times tables. (10*0, 10*1, ..., 10*12). (Use a for loop when implementing this functionality).

```python
def times_table():
    for x in range(0,12):
        print(f"10*{x} = {10*x}")
        
times_table()
```

### (iv) Your objective in this question is to write a more flexible version of the program you wrote for Question 1. You should ask the user for two values. The first value, num, is the number that we will multiply by 0, 1, 2, 3, etc. The second number, limit, is the number at which we will stop multiplying.

```python
def times_table():
    first_number = int(input("Please enter time tables for printing:"))
    second_number = int(input("Please enter upper limit for multiplication:"))
    for x in range(0,second_number+1):
        print(f"{first_number}*{x} = {first_number*x}")
        
times_table()
```

### (ii) Write a program that will ask a user for how many stars to display on a single line and also how many lines of stars to display. The program should then output the resulting sequence of stars. 

```python
num_stars = int(input("How many stars do you want in each line? "))
num_rows = int(input("How many lines of stars do you want? "))
for x in range(num_rows):
    print(f"{'*'*num_stars}")
```

### (iii) Implement a function called printNumTriangle. The function should ask the user to enter a single integer. It should then print a triangle of that size specified by the integer so that each row in the triangle is made up of the integer displayed. 

```python
num_size = int(input("Please enter an integer for triangle size: "))

for x in range(1,num_size+1):
    print(f"{str(x)*x}")
```

### Write a loop that prompts the user to enter a series of pizza toppings until they enter a 'quit' value. As they enter each topping, print a message saying you’ll add that topping to their pizza.

```python
def pizza_topping():
    
    print("Welcome to MTU pizza delivery.")
    print("We are proud to serve the best pizza in Cork.")
    print("And we are proud to add any possible topping on your pizza")
    topping = []
    while True:
        topping_new = input("Please enter a topping:")
        topping.append(topping_new)
        print(f"We included the {topping_new} to your pizza")
        
        option = input("Do you wish to add any other topping? (y/n)")
        if option == "n":
            break
pizza_topping()
```
