```python
def star_pattern(n: int) -> None:
    """
    Prints a star pattern based on the given integer.
    """
    try:
        if not isinstance(n, int):
            raise TypeError("Input must be an integer.")
        
        for i in range(1, n + 1):
            print("*" * i)
        for i in range(n - 1, 0, -1):
            print("*" * i)

    except Exception as e:
        print(f"Error: {e}")
        
star_pattern(5)
```

<p>Where:</p>

    <li>s and f stands for starting and final positions</li>

<p>Import the module "Assignment2" and use the function plot3d. This function will accept your dictionary as input <strong>(5% marks)</strong>.</p>

```python
import numpy as np 
def randomWalk(steps: int) -> dict:
    """
    Generates a 3D random walk of given steps.
    """
    try:
        if not isinstance(steps, int):
            raise TypeError("Steps must be an integer.")
        # Start at origin (0,0,0)
        x = np.cumsum(np.random.choice([-1, 1], steps))
        y = np.cumsum(np.random.choice([-1, 1], steps))
        z = np.cumsum(np.random.choice([-1, 1], steps))
        return {"X": x.tolist(), "Y": y.tolist(), "Z": z.tolist()}
    except Exception as e:
        print(f"Error: {e}")

def euclidian_distance(start: tuple, end: tuple) -> float:
    """
    Calculates Euclidean distance between start and end 3D points.
    """
    try:
        if not (len(start) == len(end) == 3):
            raise ValueError("Points must have 3 coordinates.")
        return float(np.sqrt(sum((s - f) ** 2 for s, f in zip(start, end))))
    except Exception as e:
        print(f"Error: {e}")

walk = randomWalk(5)
print(walk)
```
