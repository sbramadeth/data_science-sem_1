<p>Write a calculator program using functions. The program should first ask the user for two separate numerical values. <h4>Question 4</h4> 

<p>Write a function called printReverseNumTriangle, that accepts a number as its argument and prints a triangle of the size specified by number so that each row in the triangle is made up of the row number displayed.</p>


```python
def reverse_triangle(size:int) ->str:
    """
        reverse_triangle create a reversed triangle with the row number
        as the character * number.
        Args:
            size (int): total number of rows of the Triangle
        Returns:
            n number of rows containing n number of numbers
    """
    for x in range(1,size+1):
        """
            This loop will range from 1 to the value entered by the user plus 1.
            This is necessary because we didn't start from 0.
        """
        print(f"{' '*(size-x)}{str(x)*x}")

        """
            It will then print an empty space times the number entered by the user - the value x.
            The idea is that the, if we entered the number 5, we will always have 5 characters (including whitespace)
            So, the number one will be preceeded by 4 empty spaces, 2 by 3, 3 by 2, 4 by 1 and 5 by nothing.
            Doing like this we will be able to align our triangle in a reverse mode.
        """

reverse_triangle(5)
```

<h4>Question 5</h4>

<p>Write a simple guessing game for a user. This program should initially generate a random number between 1 and 100. 

It should then repeatedly ask the user to guess the random number. 

Each time the user enters a guess the program should tell them that their guess was too high, too low or correct. 

When the user finally guesses the correct number the program should tell the user how many guesses they made before arriving at the correct number.</p>

<p>Your program should make use of the following methods:</p>
<ol>
    <li>generateRandomNumber. This function will generate a random number between 0-100 and return the result.</li>
    <li>askUser. This function will ask the user to enter a guess and will return the result</li>
    <li>checkGuess. This function will take in the users guess and the random number as parameters and will return True if the user entered the correct value and False otherwise. </li>
</ol>

```python
import random


def genRandomNumber():
    """
        genRandomNumber will generate a random number ranging from 0 to 100
        Returns:
            A random number
    """
    return random.randint(0, 100)
    #This function will return a random number ranging from 1 to 100

def askUser():
    """
        askUser will ask the user for their guess.
        Returns:
            number (int): The user's guess
    """
    number = int(input("Please guesss a number between 0 and 100"))
    return number
    #This function will ask for an input from the user and transform it to an integer
    
def checkGuess(userGuess, randomNum):
    """
        checkGuess is a function that will be used to check if the user guessed right
        Args:
            userGuess (int): the user's guess
            randomNum (int): the random number generated by the game
        Returns:
            True: if the user guessed right
            False: if the user guessed wrong
    """
    if userGuess == randomNum:
        return True
    elif userGuess > randomNum:
        print ("Your guess is too high")
        return False
    else:
        print ("Your guess is too low")
        return False



def main():
    
    correctGuess = False
    numGuesses = 0
    
    randomNum = genRandomNumber()
    
    while correctGuess == False:
    
        guess = askUser()
        correctGuess = checkGuess(guess, randomNum)
        
        numGuesses += 1
    
    print ("Congratulations. This is the correct guess")
    print ("You made a total of ", numGuesses, " guesses")

main()
```
